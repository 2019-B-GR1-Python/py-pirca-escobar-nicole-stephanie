1. **Introducción** 

La Organización Mundial de la Salud (OMS) considera el suicidio como un problema grave de

perduran aún mucho tiempo después de la pérdida. Dentro de las causas que lo provocan se
indican diversos factores biológicos, psicológicos, sociales, ambientales y culturales.

2. **Propósito del Proyecto**

El próposito de este proyecto es brindar y analizar informacion a OMS y tomar conciencia del grave problema que esta 
afectando a nuestra sociedad definiendo que parte de la población es más vulnerable.

###  Países con el mayor índice de suicidios 
En nuestro país el fenómeno del suicidio consumado es más frecuente en la población masculina.
1.162 ciudadanos se quitaron la vida de enero a diciembre de 2017 en el país. 
De enero a agosto de este año hubo 754 suicidas lo que se traduce en que, de cada diez suicidios,

```python
plt.title('Paises Tasa de suicidios')
paises_count = data_frame_suicidios['country'].value_counts()[:].sort_values().plot(kind='barh')
count_paises = data_frame_suicidios['country'].value_counts()[:]
plt.savefig("paises_suisidio.jpg")
plt.show()
```

### Género y suicidio 
En las Américas, al igual que en otras partes del mundo, las tasas de suicidio en hombres
continúan siendo superiores a las observadas en las mujeres, y representan alrededor del 79% de
todas las defunciones por suicidio. En promedio, la razón hombre-mujer, ajustada según la edad, era
de 3,8 en la Región y 4,0 en América Latina. 

```python
plt.title('Genero vs poblacion')
genero_paises = data_frame_suicidios['suicides/100k pop'].value_counts().groupby(data_frame_suicidios['sex']).mean().plot(kind='bar')
plt.savefig("generos_porcentaje.jpg")
plt.show()
```
