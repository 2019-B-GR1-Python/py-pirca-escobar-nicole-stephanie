### Televisores

```python
Total_televisores=df_lapto['categoria']=='Televisores'
df_lapto[Total_televisores].sort_values(by=['precio_regular']).head(10)
```

```python
plt.title('Negocios por Categoria')
categoriaNegocio = df_lapto['categoria'].value_counts()[:].sort_values().plot(kind='barh')
count_categoriaNegocio = df_lapto['categoria'].value_counts()[:]
plt.savefig("negocioCategoria.jpg")
plt.show()
```

### Celulares

```python
scategoriaTelevisor=df_lapto['categoria']=='celulares'
df_lapto[scategoriaTelevisor].sort_values(by=['precio_regular']).head(15)
```

```python
plt.figure(figsize=(30,10))
ax = plt.gca()
plt.title("rating_count_tot vs rating_count_ver")
df_lapto.head(200).plot(kind='line',x='nombre',y='precio_oferta',ax=ax)
df_lapto.head(200).plot(kind='line',x='nombre',y='precio_regular', color='red', ax=ax)
plt.show()
```

### Descuentos 

```python
porcentaje_ofertas=((df_lapto.precio_regular.values-df_lapto.precio_oferta.values)/df_lapto.precio_regular.values)*100
df_lapto['descuento']=porcentaje_ofertas
df_lapto.sort_values(by=['descuento'],ascending=False).head(20)
```

```python
empresaDescuentosPrecios = sns.barplot(x='empresa', y='descuento', hue='categoria', data=df_lapto)
plt.savefig('EmpresaDescuento.jpg')
```
